```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import numpy as np
import pandas as pd
import math
```

```python
# Load data generated by C++ script
df = pd.read_csv('./build/temp.txt', sep='\t')

# Remove last row containing only prediction
df.drop(df.tail(1).index, inplace=True)
```

```python
init_notebook_mode(connected=True)
```

```python
data = [
    go.Scatter(
        x=df['px_state'],
        y=df['py_state'],
        name='UKF',
        line = dict(
            width = 5
        )
    ),
    go.Scatter(
        x=df['px_ground_truth'],
        y=df['py_ground_truth'],
        name='Ground truth',
        line = dict(
            width = 3
        )
    ),
    go.Scatter(
        x=df['px_measured'],
        y=df['py_measured'],
        name='Sensor measurement',
        line = dict(
            width = 1
        )
    )
]

layout = go.Layout(
    title='Position',
    yaxis=dict(title='P_y'),
    xaxis=dict(title='P_x')
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
df['norm_vel_ground_truth'] = df.apply(lambda row: math.hypot(row['vx_ground_truth'], row['vy_ground_truth']) , axis=1)

data = [
    go.Scatter(
        x=df.index,
        y=df['v_state'],
        name='UKF',
        line = dict(
            width = 5
        )
    ),
    go.Scatter(
        x=df.index,
        y=df['norm_vel_ground_truth'],
        name='Ground truth',
        line = dict(
            width = 3
        )
    )
]
layout = go.Layout(
    title='Velocity',
    yaxis=dict(title='Velocity'),
    xaxis=dict(title='Time')
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
